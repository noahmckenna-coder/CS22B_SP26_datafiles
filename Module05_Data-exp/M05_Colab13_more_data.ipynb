## Filtering DataFrame with string datatype
You can search for instances of a substring in a column using str.contains(), `df[colname].str.contain(<str>)`. The method will return table that meets the condition. If you want to return a single column you can define which column to print `df[df[colname].str.contain(<str>)][colname2]`.

### Series to list
To convert a Series to a list use `series.tolist()`

```python
### You can search for instances of a substring in a column and return rows
## Search for substring (str.contains) and return the table
movies[movies['Title'].str.contains("man")]

## You can print just the column 'Title'
movies[movies['Title'].str.contains("man")][['Title']]
## To convert this Series to a list use series.tolist()
man_list = movies[movies['Title'].str.contains("man")]['Title'].tolist()
man_list
```

### Question 13.1: How many distinct movies came out in the 1990s?
To find the number of distinct movies during these years:
1. Filter out the DataFrame based on conditions of the year
2. Get the length `len()` (number of rows) of your filtered DataFrame
  
-or-   
1. Use the str approach to find all years in 1990s (ie., "199")

```python
### Filter out the DataFrame based on conditions of the year combine with len
### Your code
len(movies[(movies['Year'] >= 1990) & (movies['Year'] <= 1999)])
```

```python
### Use the str approach to find all years in 1990s
movies[movies['Year'].astype(str).str.contains("199")]
```

### Question 13.2: How many distinct movies came out for each decade?
Can you write a for loop to iterate through all the decades?

```python
decades = ['191', '192', '193', '194', '195', '196', '197', '198', '199', '200', '201', '202']

for decade_prefix in decades:
    decade_movies = movies[movies['Year'].astype(str).str.startswith(decade_prefix)]
    print(f"Decade {decade_prefix}0s: {len(decade_movies)} distinct movies")
```
